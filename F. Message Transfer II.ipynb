```python
n=int(input())
for i in range(n):
    st=input()
    new_st=''
    i=0
    while(i<len(st)):
        print(i)
        if(i<len(st)-2):
            if(st[i].lower()=='w' and st[i+1].lower()=='u' and st[i+2].lower()=='b'):
                print('hi1')
                i+=3
            else:
                while( (st[i].lower()!='w' or st[i+1].lower()!='u' or st[i+2].lower()!='b') and i<len(st)):
                    print('hi2')
                    new_st+=st[i]
                    i+=1
                new_st+=' '
                #i+=1
        else:
            print('hi3')
            new_st+=st[i]
            i+=1
        print("&&",new_st)
    print("**",new_st)
```
