```python
_=int(input())
for _ in range(_):
    st=input()
    di={}
    for i in range(len(st)):
        if(st[i] not in di):
            di[st[i]]=1
        else:
            di[st[i]]+=1
    cnt=1
    for k,v in di.items():
        cnt=cnt*v
    print(di)
    print(cnt)
            
```
