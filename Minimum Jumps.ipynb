```python
def explore(current,final,k1,k2,last,arr,jumps):
    print(current,final,k1,k2,last,jumps)
    if(current==final):
        print("1")
        return jumps
    elif(current in arr):
        print("2")
        return 0
    elif(current>final):
        if(last=='b'):
            print("3")
            return 0
        else:
            print("4")
            last='b'
            explore(current-k2,final,k1,k2,last,arr,jumps+1)
    elif(current<final):
        print("5")
        last='f'
        l=explore(current+k1,final,k1,k2,last,arr,jumps+1)
        last='b'
        r=explore(current-k2,final,k1,k2,last,arr,jumps+1)
        return l+r 
```
