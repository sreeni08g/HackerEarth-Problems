```python
def xnor(a,b,length):
    result =[]
    binary_a =list(bin(a)[2:])
    binary_b =list(bin(b)[2:])
    #print(binary_a,binary_b)
    if(len(binary_a)<length):
        diff = length-len(binary_a)
        for i in range(diff):
            binary_a.insert(0,'0')
    if(len(binary_b)<length):
        diff = length-len(binary_b)
        for i in range(diff):
            binary_b.insert(0,'0')
    #print(binary_a,binary_b)
    if(len(binary_a)==len(binary_b)):
        for i in range(length):
            if(binary_a[i]=='1' and binary_b[i]=='1'):
                result.append('1')
            elif(binary_a[i]=='0' and binary_b[i]=='0'):
                result.append('1')
            else:
                result.append('0')
    #print(result)
    result.insert(0,'b')
    result.insert(0,'0')
    #print(result)
    return(int(''.join(result),base=0))
            
        
    
```

```python
xnor(4,5,3)
```

```python
di_xoro={1:1,2:2}
def xoro(n):
    print(n)
    if (n==0):
        return 0
    elif(n in di_xoro):
        return di_xoro[n]
    else:
        di_xoro[n]=xoro(n-1)^xoro(n-2)
        return di_xoro[n]
```
